```python
tickers = ['AMZN','NFLX','GOOG','MSFT']
tickers
```

```python
tickers = [yf.Ticker(ticker) for ticker in tickers]
tickers
```

```python
#dataframes
dfs = []
for ticker in tickers:
    # get financial statements for each
    pnl = ticker.financials
    bs = ticker.balancesheet
    cf = ticker.cashflow
    
    # concat into one dataframe
    fs = pd.concat([pnl, cf])
    
    # make dataframe format nicer
    data = fs.T # swap data and columns
    data = data.reset_index() # reste index (data) into a column
    data.columns = ['Date', *data.columns[1:]] # rename od index to Date
    data['Ticker'] = ticker.ticker # add ticker to dataframe
    dfs.append(data)
data
```

```python
parser = pd.io.parsers.base_parser.ParserBase({'usecols': None})

for df in dfs:
    df.columns = parser._maybe_dedup_names(df.columns)
    
df = pd.concat(dfs, ignore_index=True)
df = df.set_index(['Ticker','Date'])
df
```
